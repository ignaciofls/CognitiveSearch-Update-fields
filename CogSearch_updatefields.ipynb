```python
def get_all_docs():

    facet = 'organizations'
    documents = []
    searchstring = '&search=*&$count=true&$select=locations,metadata_storage_path'
    url = endpoint + "indexes/" + indexname +"/docs" + api_version + searchstring
    response  = requests.get(url, headers=headers, json=searchstring)
    query = response.json()
    #pprint(url)

    for doc in query['value']:
        documents.append(doc)

        # Continue if needed
        while('@odata.nextLink' in query.keys()):
            next_link = query['@odata.nextLink']
            #pprint(next_link)

            response = requests.get(next_link, headers=headers)
            query = response.json()

            for doc in query['value']:
                documents.append(doc)

            #pprint(query)
            
    return documents
```

```python
documents = get_all_docs()
```

# Push data to index

```python
#First we format the dump to remove the search_score property, change field to new name
search_docs = {
    "value" : []
}
documents = get_all_docs()

for d in documents:
    del d['@search.score'] 
    d['@search.action'] = 'mergeOrUpload'
    search_docs['value'].append(d)
```

```python
with open('temp.txt', 'w') as json_file:
  json.dump(search_docs, json_file)
```
